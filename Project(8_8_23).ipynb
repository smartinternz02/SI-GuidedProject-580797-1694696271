```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
crashes = sns.load_dataset('car_crashes')
print(crashes)


```

```python
print(crashes.head())
```

```python
print(crashes.tail())
```

```python
sns.pairplot(crashes)
plt.title("Pair Plot of Car Crashes Data")
plt.show()
```

```python
correlation_matrix = crashes.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Car Crashes Data")
plt.show()
```

```python
sns.barplot(x="abbrev", y="total", data=crashes, palette="viridis")
plt.title("Total Crashes by State")
plt.xlabel("State Abbreviation")
plt.ylabel("Total Crashes")
plt.show()
```

```python
sns.boxplot(x="abbrev", y="ins_premium", data=crashes, palette="Set2")
plt.title("Distribution of Insurance Premiums by State")
plt.xlabel("State Abbreviation")
plt.ylabel("Insurance Premium")
plt.show()
```

```python
sns.violinplot(x="abbrev", y="alcohol", data=crashes, palette="husl")
plt.title("Distribution of Alcohol Consumption by State")
plt.xlabel("State Abbreviation")
plt.ylabel("Alcohol Consumption")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(crashes['total'], label='Total Crashes', color='blue')
plt.title('Line Graph of Total Crashes')
plt.xlabel('Index')
plt.ylabel('Total Crashes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(crashes['not_distracted'], bins=15, edgecolor='black', alpha=0.7, color='purple')
plt.title('Histogram of Not Distracted Drivers')
plt.xlabel('Not Distracted Drivers')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
age_group_counts = crashes['alcohol'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('Pie Chart of Alcohol Percentage')
plt.show()
```

```python
sns.jointplot(x='total', y='alcohol', data=crashes, kind='scatter')  # You can use kind='scatter', 'reg', 'kde', 'hex', etc.
plt.title('Joint Plot of Total Crashes vs. Alcohol Consumption')
plt.show()
```

```python
sns.distplot(crashes['not_distracted'], bins=15, kde=False, color='blue')
plt.title('Distribution of Not Distracted Drivers')
plt.xlabel('Not Distracted Drivers')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.relplot(x='alcohol', y='total', data=crashes, hue='abbrev', palette='viridis', kind='scatter')
plt.title('Relational Plot of Alcohol vs. Total Crashes by State')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Total Crashes')
plt.show()
```
